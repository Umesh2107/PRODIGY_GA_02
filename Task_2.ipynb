```python
# 🚀 All-in-One: Install, Download, Generate Image with Stable Diffusion (No Hugging Face Login)
!pip install -q torch torchvision torchaudio diffusers transformers scipy safetensors accelerate

import os
from PIL import Image
import torch
from diffusers import StableDiffusionPipeline

# Step 1: Create model folder & download weights
model_path = "/content/stable-diffusion-v1-5"
os.makedirs(model_path, exist_ok=True)

!wget -q https://huggingface.co/runwayml/stable-diffusion-v1-5/resolve/main/v1-5-pruned-emaonly.safetensors -O {model_path}/model.safetensors

# Step 2: Load the model
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    use_safetensors=True
)
pipe.to("cuda")

# Step 3: Generate image from prompt
prompt = "a futuristic city with flying cars and neon lights"
image = pipe(prompt).images[0]

# Step 4: Display and save image
image.show()
image.save("/content/generated_image.png")
print("✅ Image saved as /content/generated_image.png")
```
